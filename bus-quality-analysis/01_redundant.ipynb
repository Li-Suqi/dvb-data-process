```python
import polars as pl

DATA_PATH = "../data/bus_2025-07-28_2025-08-03.parquet"
OUTPUT_DIR = "quality_report"

df = (
    pl.read_parquet(DATA_PATH)
    .with_columns(
        pl.col("tst_iso").str.to_datetime(format='%Y-%m-%dT%H:%M:%S%.f%z')
    )
)

print(f"总行数: {len(df):,}")
print(df.dtypes)
print(df.head(3))
```

```python
# Cell 2 — Run analysis
import matplotlib.pyplot as plt
from quality_utils import analyze_redundant
result = analyze_redundant(df, output_dir="quality_report")

print(f"Hard duplicates : {result['hard_dup_count']:,}  ({result['hard_dup_rate']:.3%})")
print(f"Soft dup groups : {result['soft_dup_affected_groups']:,}")
print(f"Soft dup extra  : {result['soft_dup_extra_rows']:,}  ({result['soft_dup_rate']:.3%})")
```
