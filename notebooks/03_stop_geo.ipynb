```python
from pathlib import Path
import polars as pl

DATA_DIR = Path("..") / "data"
RAW_CSV  = DATA_DIR / "stop_geometry.csv"
OUT_PATH = DATA_DIR / "stop_geometry.parquet"
```

```python
df = pl.read_csv(RAW_CSV)
print(f"Loaded:  {len(df):,} rows")
print(df.schema)
df.head(5)
```

```python
df_unique = df.unique(maintain_order=True)
print(f"Before: {len(df):,}  |  After dedup: {len(df_unique):,}  |  Removed: {len(df) - len(df_unique):,}")
df_unique.head(5)
```

```python
df_unique.write_parquet(OUT_PATH)
print(f"Saved → {OUT_PATH}  ({OUT_PATH.stat().st_size / 1e3:.1f} KB)")
```
